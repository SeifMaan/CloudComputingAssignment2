```python
import pandas as pd

books_df = pd.read_csv(r'books.csv')
print(books_df.head())

print("Missing values before cleaning:")
print(books_df.isnull().sum())

books_df.dropna(inplace=True)

#Convert letters to lower case
books_df['clean_text'] = books_df['title'].str.lower()

#Remove spaces
books_df['clean_text'] = books_df['clean_text'].str.strip()

#Cleaned Data
print(books_df.head())

harry_potter_df = books_df[books_df['title'].str.contains('harry potter', case=False)]

most_selling_books = harry_potter_df.sort_values(by='ratings_count', ascending=False)

print("Most Selling Books within the Harry Potter Series:")
print(most_selling_books[['title', 'ratings_count']].head())

average_rating = harry_potter_df['average_rating'].mean()

print("\nAverage Rating of the Harry Potter Books:", average_rating)
```
